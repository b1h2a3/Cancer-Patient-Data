```python
# ploting the Age of patients in cancer

ax = sns.countplot(x='AGE', data=df)
#sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# ploting pateint count as gender 
ax = sns.countplot(x='GENDER', data=df)
#sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

It means males are the most cancer patient 

```python
# ploting number of smoking person 
ax = sns.countplot(x='SMOKING', data=df)
#sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

Most people are cancer patient who smokes

```python
# ploting number of chest pain patient
ax = sns.countplot(x='CHEST_PAIN', data=df)
#sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

Almost half patient have chest pain
